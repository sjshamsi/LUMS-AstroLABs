```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('seaborn')
plt.style.use('seaborn-colorblind')
```

```python
data = pd.read_excel('AM0AM1_5.xls', header=1)
data = data[data['Wavelength (nm)'] < 1400]
data
```

```python
wavlens1 = data['Wavelength (nm)']
ext = data['Extraterrestrial W*m-2*nm-1']
global_tilt = data['Global tilt  W*m-2*nm-1']
d_s = data['Direct+circumsolar W*m-2*nm-1']

wavlens2 = data['Wavelength (nm).1']
irr2 = data['W*m-2*nm-1']
```

```python
plt.figure(figsize=(16, 10))

plt.plot(wavlens1, ext, linewidth=0.8, label='Extraterrestrial')
plt.plot(wavlens1, global_tilt, linewidth=0.8, label='Global Tilt')
plt.plot(wavlens1, d_s, linewidth=0.8, label='Direct + Circumsolar')

plt.plot(wavlens2, irr2, linewidth=0.8, label='ASTM E-490 AM0 Standard Spectra')

plt.xlim(300, 1400)

plt.xlabel('Wavelength $\\lambda$', size=20, labelpad=20)
plt.ylabel('Spectral Emissison $W m^{-2} nm^{-1}$', size=20, labelpad=20)

plt.xticks(np.arange(300, 1400, 50), size=15)
plt.yticks(size=15)

plt.legend(fontsize=16)
```
