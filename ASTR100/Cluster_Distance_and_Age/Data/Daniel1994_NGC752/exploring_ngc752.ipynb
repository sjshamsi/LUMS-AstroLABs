```python
from astropy.io import fits

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stars_fits = fits.open('daniel1994_ngc752_photometry.fit')
stars = pd.DataFrame(stars_fits[1].data)

ngc752_e = 0.03
stars['B-V_0'] = stars['B-V'] - - ngc752_e

stars
```

```python
stars_filt = stars[stars['PlatPR'] > .5]
```

### Let's try to add a V Mag slider

```python
from matplotlib.widgets import Slider, Button
```

```python
%matplotlib widget

fig = plt.figure()
ax = fig.subplots()

plt.subplots_adjust(bottom=0.25)

ax.scatter(stars_filt['B-V_0'], stars_filt['Vmag'], s=2)
p, = ax.plot(B-V, V)

plt.ylim(0, 15)
plt.xlim(0, 1.5)
plt.gca().invert_yaxis()

sliderax = plt.axes([0.25, 0.1, 0.65, 0.03])

d = Slider(sliderax, 'Distance', valmin=-10, valmax=3, valinit=0, valstep=0.1)

def update(val):
    diff = d.val
    adj_V = V - diff
    p.set_ydata(adj_V)
    fig.canvas.draw_idle()
    
d.on_changed(update)

resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', color='violet', hovercolor='purple')

def resetSlider(event):
    d.reset()

button.on_clicked(resetSlider)

# plt.show()
```

### Great success! Now we can try fitting the isochrone w.r.t year and hope we can determine the age of the cluster!

```python
%matplotlib widget

fig = plt.figure(figsize=(5, 5))
ax = fig.subplots()

plt.subplots_adjust(bottom=0.35)

minage = min(isocmd.ages)
maxage = max(isocmd.ages)

age_ind = isocmd.age_index(minage)
B = isocmd.isocmds[age_ind]['Bessell_B']
V = isocmd.isocmds[age_ind]['Bessell_V']
p, = ax.plot(B - V, V)

ax.scatter(stars_filt['B-V_0'], stars_filt['Vmag'], s=2)

plt.ylim(6, 16)
plt.xlim(0, 1.5)
plt.gca().invert_yaxis()

dax = plt.axes([0.2, 0.2, 0.6, 0.03])
ageax = plt.axes([0.2, 0.1, 0.6, 0.03])

d_slider = Slider(dax, 'Distance', valmin=-10, valmax=3, valinit=0, valstep=0.05)
age_slider = Slider(ageax, 'Age', valmin=minage, valmax=maxage, valinit=minage, valstep=0.05)

def update(val):
    age_ind = isocmd.age_index(age_slider.val)
    B = isocmd.isocmds[age_ind]['Tycho_B']
    V = isocmd.isocmds[age_ind]['Tycho_V']
    
    p.set_xdata(B-V)
    p.set_ydata(V - d_slider.val)
    fig.canvas.draw_idle()
    
d_slider.on_changed(update)
age_slider.on_changed(update)

reset_d = plt.axes([0.85, 0.2, 0.1, 0.03])
reset_age = plt.axes([0.85, 0.1, 0.1, 0.03])

d_button = Button(reset_d, 'Reset', color='violet', hovercolor='purple')
age_button = Button(reset_age, 'Reset', color='violet', hovercolor='purple')

def resetDSlider(event):
    d_slider.reset()
    
def resetAgeSlider(event):
    age_slider.reset()

d_button.on_clicked(resetDSlider)
age_button.on_clicked(resetAgeSlider)
```
